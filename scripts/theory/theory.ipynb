```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
def extraMemory(sigma):
  return 1 - min(2/3, 32/(3 * sigma))
  
```

```python
sigma = list(range(1, 100))
emem = [extraMemory(s) for s in sigma]
```

```python
plt.plot(sigma, emem)
plt.hlines(0.5, 0, 100, colors='red')
plt.hlines(0.5, 0, 100, colors='red')
plt.ylim([0, 1])
```
